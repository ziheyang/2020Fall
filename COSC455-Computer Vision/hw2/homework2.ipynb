```python
# question 2
import matplotlib.pyplot as plt
from skimage import io
import numpy as np
import os
%matplotlib inline

#path of images 
path = os.path.expanduser("imagesHm2")
bottle = os.path.join(path, "drip-bottle.tif")
iris = os.path.join(path, "iris-RGB.tif")

def get_color_channel(img, idx):
    # extract color channel
    color_channel = img[:,:,idx]

    # create empty image with same shape as the source image
    color_img = np.zeros(img.shape, dtype="uint8")

    # assign the color channel of source to empty image
    color_img[:,:,idx] = color_channel
    
    return color_img

img2 = io.imread(iris)

#create a figure
fig2 = plt.figure(num=1, figsize=(12, 19))

#create subplots for fig
rows = 1
cols = 4
myplots = fig2.subplots(rows, cols)

#show image
myplots[0].imshow(img2)
myplots[0].axis('off')

img_r = get_color_channel(img2, 0)
myplots[1].imshow(img_r)
myplots[1].axis('off')

img_g = get_color_channel(img2, 1)
myplots[2].imshow(img_g)
myplots[2].axis('off')

img_b = get_color_channel(img2, 2)
myplots[3].imshow(img_b)
myplots[3].axis('off')
```
