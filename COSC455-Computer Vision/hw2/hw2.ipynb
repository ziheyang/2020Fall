```python
import matplotlib.pyplot as plt
from skimage import io
import numpy as np

import os

#path of images 
path = os.path.expanduser("imagesHm2")
bottle = os.path.join(path, "drip-bottle.tif")
iris = os.path.join(path, "iris-RGB.tif")

#read iamges
img1 = io.imread(bottle)

#create a figure
fig = plt.figure(num=1, figsize=(10, 10))

#create subplots for fig
rows = 2
cols = 4
myplots = fig.subplots(rows, cols)

#show image
myplots[0][0].imshow(img1)
myplots[0][0].axis('off')

myplots[0][1].imshow(img1)
myplots[0][1].axis('off')

myplots[0][2].imshow(img1)
myplots[0][2].axis('off')

myplots[0][3].imshow(img1)
myplots[0][3].axis('off')

myplots[1][0].imshow(img1, cmap = 'gray')
myplots[1][0].axis('off')

myplots[1][1].imshow(img1, cmap = 'gray')
myplots[1][1].axis('off')

myplots[1][2].imshow(img1, cmap = 'gray')
myplots[1][2].axis('off')

k = pow(2, 7)
img3 = img1 / k
myplots[1][3].imshow(img3, cmap = 'gray')
myplots[1][3].axis('off')
```

```python
import cv2
import numpy as np
# Load the image
img = cv2.imread('imagesHm2/drip-bottle.tif',0)
# Find width and height of image
row, column = img.shape
# Create an zeros array to store the sliced image
img1 = np.zeros(img.shape,dtype = 'uint8')

# Specify the min and max range
min_range = 0
max_range = 16

# Loop over the input image and if pixel value lies in desired range set it to 255 otherwise set it to 0.
for i in range(row):
    for j in range(column):
        if img[i,j]>min_range and img[i,j]<max_range:
            img1[i,j] = img[i,j]
        else:
            img1[i,j] = 0
# Display the image
cv2.imshow('sliced image', img1)
```

```python
%matplotlib inline

def get_color_channel(img, idx):
    # extract color channel
    color_channel = img[:,:,idx]

    # create empty image with same shape as the source image
    color_img = np.zeros(img.shape, dtype="uint8")

    # assign the color channel of source to empty image
    color_img[:,:,idx] = color_channel
    
    return color_img

img2 = io.imread(iris)

#create a figure
fig2 = plt.figure(num=1, figsize=(12, 19))

#create subplots for fig
rows = 1
cols = 4
myplots = fig2.subplots(rows, cols)

#show image
myplots[0].imshow(img2)
myplots[0].axis('off')

img_r = get_color_channel(img2, 0)
myplots[1].imshow(img_r)
myplots[1].axis('off')

img_g = get_color_channel(img2, 1)
myplots[2].imshow(img_g)
myplots[2].axis('off')

img_b = get_color_channel(img2, 2)
myplots[3].imshow(img_b)
myplots[3].axis('off')
```
